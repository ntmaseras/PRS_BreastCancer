```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("simulation/taggings/her.tagging", sep=" ")

df['Exp_Heritability'] = pd.to_numeric(df['Exp_Heritability'], errors='coerce')
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Exp_Heritability'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Exp_Heritability')
plt.ylabel('Frequency')
plt.title('Distribution of Exp_Heritability in the her.tagging')
plt.grid(axis='y')

# Show the plot
plt.show()
plt.savefig('results/Exp_heritability.png')
```

```python
df = df[df['Exp_Heritability'] > 0.4]
plt.figure(figsize=(12, 6))
plt.scatter(df["Predictor"], df["Exp_Heritability"], color='skyblue', marker='o')
plt.xlabel('Predictor')
plt.ylabel('Exp_Heritability')
plt.title('Expected Heritability for All SNPs')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.tight_layout()

# Save the plot as an image file before displaying it
plt.savefig('results/HERITABILITY.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
